## 1. Pytesseract for OCR

```python
# Import image
img = cv2.imread('limerick.PNG')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Show image
plt.imshow(img)
```

```python
# Detect and print words contained in the image
print(pytesseract.image_to_string(img))
```

```python
# Bounding box for word
# By printing this, you get the x,y,w,h coordinates of the boxes
print(pytesseract.image_to_boxes(img))
```

```python
# Draw bounding boxes

# Import image
img = cv2.imread('limerick.PNG')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

h_img, w_img, no_channel = img.shape
boxes = pytesseract.image_to_boxes(img)

for b in boxes.splitlines():
    #print(b)
    b = b.split(' ')
    #print(b)
    x,y,w,h = int(b[1]), int(b[2]), int(b[3]), int(b[4])
    cv2.rectangle(img, (x, h_img-y), (w,h_img-h), (255,0,0), 1)
    cv2.putText(img, b[0], (x, h_img-y+10), cv2.FONT_HERSHEY_COMPLEX,0.4, (0,0,0),1)

#plt.clf()
plt.figure(figsize = (20,10))
plt.imshow(img)
```
